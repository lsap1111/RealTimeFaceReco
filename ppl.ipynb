```python
import cv2
import os

# Path to the folder where authorized faces are saved
authorized_faces_folder = r"C:\Users\hp\Downloads\project\Untitled Folder 3\authorized_faces"

# Load the Haar Cascade model for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize face recognizer (LBPH Face Recognizer)
recognizer = cv2.face.LBPHFaceRecognizer_create()

# Load the authorized face for recognition
authorized_face_path = os.path.join(authorized_faces_folder, "authorized_face.jpg")
authorized_face_img = cv2.imread(authorized_face_path)

if authorized_face_img is None:
    print("‚ùå No authorized face found! Please capture an authorized face first.")
    exit()

# Convert the authorized face image to grayscale for face recognition
authorized_face_gray = cv2.cvtColor(authorized_face_img, cv2.COLOR_BGR2GRAY)

# Train the recognizer with the authorized face
recognizer.train([authorized_face_gray], np.array([0]))

# Start webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("‚ùå Error: Could not open webcam.")
    exit()

print("üì∏ Press 'q' to quit.")

while True:
    ret, frame = cap.read()

    if not ret:
        print("‚ùå Failed to grab frame")
        break

    # Convert the frame to grayscale for face detection
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5)

    # For each detected face, recognize and draw a rectangle around it
    for (x, y, w, h) in faces:
        detected_face = gray_frame[y:y + h, x:x + w]

        # Predict the label for the detected face (whether it matches the authorized face)
        label, confidence = recognizer.predict(detected_face)

        # If the confidence is below a threshold (i.e., recognized as the authorized face)
        if confidence < 100:
            cv2.putText(frame, f"Authorized - {round(100 - confidence)}%", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        else:
            cv2.putText(frame, "Unauthorized", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

    # Show the frame with face detection and recognition result
    cv2.imshow("Real-Time Face Recognition", frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os
import numpy as np

data_path = r"C:\Users\hp\Downloads\project\Untitled Folder 3\authorized_faces"
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
recognizer = cv2.face.LBPHFaceRecognizer_create()

faces = []
labels = []
label_map = {}
current_id = 0

for person_name in os.listdir(data_path):
    person_path = os.path.join(data_path, person_name)
    if not os.path.isdir(person_path):
        continue

    label_map[current_id] = person_name

    for img_name in os.listdir(person_path):
        img_path = os.path.join(person_path, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue

        faces.append(img)
        labels.append(current_id)
    current_id += 1

recognizer.train(faces, np.array(labels))
print("‚úÖ Training completed with", len(label_map), "authorized users.")
```

```python
import threading
import cv2
import os
import numpy as np
import datetime
import time


# Path to the folder where authorized faces are saved
authorized_faces_folder = r"C:\Users\hp\Downloads\project\Untitled Folder 3\authorized_faces"

# Load Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize LBPH face recognizer
recognizer = cv2.face.LBPHFaceRecognizer_create()

# Threshold for confidence (lower means more confident)
CONFIDENCE_THRESHOLD = 100

# Map label IDs to user names (folder names)
def get_label_name_map(authorized_faces_folder):
    label_name_map = {}
    for idx, person_name in enumerate(os.listdir(authorized_faces_folder)):
        person_path = os.path.join(authorized_faces_folder, person_name)
        if os.path.isdir(person_path):
            label_name_map[idx] = person_name
    return label_name_map

label_name_map = get_label_name_map(authorized_faces_folder)

# Function to load faces and labels from authorized_faces_folder
def load_authorized_faces(authorized_faces_folder):
    images = []
    labels = []
    label = 0
    for person_name in os.listdir(authorized_faces_folder):
        person_folder = os.path.join(authorized_faces_folder, person_name)
        if not os.path.isdir(person_folder):
            continue
        for file in os.listdir(person_folder):
            if file.lower().endswith((".jpg", ".png")):
                image_path = os.path.join(person_folder, file)
                img = cv2.imread(image_path)
                if img is None:
                    print(f"Warning: Could not read image {image_path}")
                    continue
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
                for (x, y, w, h) in faces:
                    face = gray[y:y + h, x:x + w]
                    images.append(face)
                    labels.append(label)
        label += 1
    return images, labels

print("üìÇ Loading authorized faces...")
faces, labels = load_authorized_faces(authorized_faces_folder)

if len(faces) == 0:
    print("‚ùå No authorized faces found! Please add authorized faces in folders.")
    exit()

# Train recognizer
recognizer.train(faces, np.array(labels))

# Create folder for unauthorized snapshots if it doesn't exist
unauthorized_folder = os.path.join(os.getcwd(), "unauthorized_snapshots")
if not os.path.exists(unauthorized_folder):
    os.makedirs(unauthorized_folder)

# Email cooldown dictionary and period (in seconds)
email_cooldown = {}
COOLDOWN_PERIOD = 300  # 5 minutes


import smtplib
from email.message import EmailMessage





from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import os

# Email details
sender_email = "sender@gmail.com"
receiver_email = "receiver@gmail.com"
password = "password"  # Use App Password for Gmail

# SMTP server details
smtp_server = "smtp.gmail.com"
smtp_port = 465

def send_email(subject, body, attachment_path=None):
    message = MIMEMultipart()
    message["From"] = sender_email
    message["To"] = receiver_email
    message["Subject"] = subject

    # Attach body
    message.attach(MIMEText(body, "plain"))

    # Attach image if provided
    if attachment_path and os.path.exists(attachment_path):
        with open(attachment_path, "rb") as attachment:
            part = MIMEBase("application", "octet-stream")
            part.set_payload(attachment.read())
            encoders.encode_base64(part)
            part.add_header(
                "Content-Disposition",
                f"attachment; filename={os.path.basename(attachment_path)}",
            )
            message.attach(part)

    # Send email
    context = ssl.create_default_context()
    with smtplib.SMTP_SSL(smtp_server, smtp_port, context=context) as server:
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, message.as_string())

    print("‚úÖ Email sent with attachment!")




def send_email_threaded_with_cooldown(image_path):
    current_time = time.time()
    key = "unauthorized"  # Apply cooldown globally for all unauthorized faces
    last_sent_time = email_cooldown.get(key, 0)
    
    if current_time - last_sent_time > COOLDOWN_PERIOD:
        email_cooldown[key] = current_time
        thread = threading.Thread(target=send_email, args=(
            "üö® Unauthorized Face Detected",
            "An unauthorized face was detected. See attached snapshot.",
            image_path
        ))
        thread.daemon = True
        thread.start()
    else:
        print(f"‚ö†Ô∏è Email cooldown active, skipping send.")


# Start webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("‚ùå Error: Could not open webcam.")
    exit()

print("üì∏ Press 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ùå Failed to grab frame")
        break

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces_in_frame = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces_in_frame:
        detected_face = gray_frame[y:y + h, x:x + w]
        try:
            label, confidence = recognizer.predict(detected_face)
        except Exception as e:
            print(f"Recognition error: {e}")
            continue

        if confidence < CONFIDENCE_THRESHOLD:
            user_name = label_name_map.get(label, f"User {label}")
            cv2.putText(frame, f"{user_name} - {round(100 - confidence)}%", (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        else:
            # Unauthorized face detected - save snapshot
            cv2.putText(frame, "Unauthorized", (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

            face_img = frame[y:y + h, x:x + w]
            timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
            file_path = os.path.join(unauthorized_folder, f"unauth_{timestamp}.jpg")
            cv2.imwrite(file_path, face_img)

            # Send email alert with cooldown
            send_email_threaded_with_cooldown(file_path)

    cv2.imshow("Real-Time Face Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
